##### Grouping of 'Aggregate rating', 'Rating color', 'Rating text'

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index() #as in dataframe
```

```python
#Add one more column named 'Rating count' after observing there is no specific column for the count 
#Assign the whole group as ratings
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
ratings
```

#### Obeservations of Ratings
1. Excellent - when 4.5 < rating < 4.9
2. Very Good - when 4.0 < rating < 4.4
3. Good      - when 3.5 < rating < 3.9
4. Average   - when 2.5 < rating < 3.4
5. Poor      - when 1.8 < rating < 2.4
6. Not Rated - when 0.0

```python
#Bar Plot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data = ratings) 
```

```python
sns.barplot(x = 'Aggregate rating', y='Rating count', hue='Rating color', data = ratings, palette =['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

#### Observation
1. Not Rated count is very high
2. Max no. of rating are between 2.5 and 3.4

```python
#Count Plot
sns.countplot(x='Rating color', data = ratings, palette= ['Blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```
